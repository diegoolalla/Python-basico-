```python
import matplotlib.pyplot as plt
import pandas as pd
import random
import os
import getpass

def submenu(opcion_principal):
    intentos = {1: 20, 2: 12, 3: 5}
    while True:
        print(f"\nSubmenú de la opción {opcion_principal}")
        print("1. Facil (20 intentos)")
        print("2. Medio (12 intentos)")
        print("3. Dificil (5 intentos)")
        try:

            opcion_submenu = int(input("Selecciona una opción: "))
            if opcion_submenu == 1:
                tipo_juego(opcion_principal, intentos[1])
                break
            elif opcion_submenu == 2:
                tipo_juego(opcion_principal, intentos[2])
                break
            elif opcion_submenu == 3:
                tipo_juego(opcion_principal, intentos[3])
                break
            else:
                print("Opción no válida. Inténtalo de nuevo.")
        except ValueError:
            print('El valor introducido no es correcto')
```

```python
def menu():
    while True:
        print("\nMenú Principal")
        print("1. Partida modo solitario")
        print("2. Partida 2 jugadores")
        print("3. Estadística")
        print("4. Salir")
        try:

            opcion = int(input("Selecciona una opción: "))
            if opcion == 1:
                submenu(opcion)
            elif opcion == 2:
                submenu(opcion)
            elif opcion == 3:
                mostrar_estadistica()
            elif opcion == 4:
                print("Saliendo del programa. ¡Hasta luego!")
                break
            else:
                print("Opción no válida. Inténtalo de nuevo.")
        except ValueError:
            print('El valor introducido no es correcto')
```

```python
def tipo_juego(tipo_de_juego, intentos):
    if tipo_de_juego == 1:
        print("has elegido partida modo solitario")
        solitario(intentos)
    else:
        print('has elegido partida dos jugadores')
        dos_jugadores(intentos)
```

```python
def solitario(intentos):
    numero = random.randint(1, 1000)
    juego(intentos, numero)

def dos_jugadores(intentos):
    while True:
        try:
            numero_cifrado = int(getpass.getpass('Jugador 1 introduce un número para que lo adivine jugador 2: '))
            if 1 <= numero_cifrado <= 1000:
                juego(intentos, numero_cifrado)
                break
            else:
                print('El número debe estar entre 1 y 1000.')
        except ValueError:
            print('El valor introducido no es correcto.')

```

```python
def juego(intentos, numero_adivinar):
    resultado = False
    intentos_restantes = intentos
    
    while intentos_restantes > 0:
        print(f'Intento {intentos - intentos_restantes + 1} de {intentos}.')
        try:
            numero_usuario = int(input('Adivina el número (entre 1 y 1000): '))
            
            if 1 <= numero_usuario <= 1000:
                intentos_restantes -= 1  
                if numero_usuario == numero_adivinar:
                    print('¡Felicidades, has adivinado el número!')
                    resultado = True
                    break
                elif numero_usuario < numero_adivinar:
                    print('El número secreto es mayor que el número introducido.')
                else:
                    print('El número secreto es menor que el número introducido.')
            else:
                print('El número debe estar entre 1 y 1000. Este intento no cuenta.')
                
        except ValueError:
            print('El valor introducido no es correcto.')

    if not resultado:
        print(f'Has agotado el número de intentos. El número secreto era {numero_adivinar}.')
    
    guardar_estadistica(resultado)
```

```python
def guardar_estadistica(resultado):
    while True:
        nombre = input('Introduce tu nombre: ')
        if nombre.isalpha():
            break
        else:
            print('El nombre solo puede contener letras. Inténtalo de nuevo.')

    datos = {
        "resultado": [resultado],
        "nombre": [nombre],
    }
    df = pd.DataFrame(datos)
    
    if os.path.exists("estadistica.xlsx"):
        df_excel = pd.read_excel("estadistica.xlsx")
        df_final = pd.concat([df_excel, df], ignore_index=True)
    else:
        df_final = df
    
    df_final.to_excel("estadistica.xlsx", index=False)
    print('Partida guardada.')



def mostrar_estadistica():
    df = pd.read_excel("estadistica.xlsx")
    # Contar los resultados por persona
    conteo = df.groupby(["nombre", "resultado"]).size().unstack(fill_value=0)

    # Gráfico de barras agrupadas
    conteo.plot(kind="bar", stacked=False)
    plt.title("Resultados por jugador")
    plt.xlabel("Jugador")
    plt.ylabel("Cantidad")
    plt.legend(['Victorias','Derrotas'], title='Resultado')
    plt.show()

menu()
```
